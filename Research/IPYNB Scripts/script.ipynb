```python
import pandas as pd
import ast

file_path = 'nasdaq100_data.csv'

df = pd.read_csv(file_path)
```

```python
data = df[['Ticker', 'Articles']]
data['Articles'] = data['Articles'].apply(lambda x: ast.literal_eval(x))
```

```python
grouped = data.groupby('Ticker')['Articles'].apply(list)

# convert the resulting DataFrame into a dictionary
result_dict = grouped.to_dict()

result_dict
```

```python
for key in result_dict.keys():
    result_dict[key] = result_dict[key][0]
```

```python
len(result_dict['MSFT'])
```

```python
news_database = []

for key in result_dict.keys():
    ticker_details = {}
    ticker_details['ticker'] = key
    ticker_details['news_items'] = []
    for item in result_dict[key]:
        news_item = {}
        news_item['title'] = item[0]
        news_item['source'] = item[1]
        ticker_details['news_items'].append(news_item)
    news_database.append(ticker_details)
news_database    

```
